```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/BRCA.csv')
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

# **As we can see all of these conatains NULL value so we have to balance it**

```python
mode_age = df['Age'].mode()
mean_age = df['Age'].mean()
median_age = df['Age'].median()
var_age = df['Age'].var()
std_age = df['Age'].std()

print(f'mode_age = ',mode_age)
print(f'mean_age = ',mean_age)
print(f'median_age = ',median_age)
print(f'var_age = ',var_age)
print(f'std_age = ',std_age)
```

# **Age value fillup with Medians**

```python
df['Age'].fillna(median_age ,inplace = True)
```

```python
mode_hist = df['Histology'].mode()


print(f'mode_hist = ',mode_hist)
```

# **Histology Null value fillp with MOde**

```python
df['Histology'].fillna(mode_hist[0],inplace = True)
```
